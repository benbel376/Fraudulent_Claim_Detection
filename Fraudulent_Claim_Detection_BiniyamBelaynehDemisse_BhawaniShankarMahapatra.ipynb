### **4.4 Perform bivariate analysis** <font color="red">[10 Marks]</font>

#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
import numpy as np
import pandas as pd
from IPython.display import display

try:
    from scipy.stats import chi2_contingency
    _HAS_SCIPY = True
except Exception:
    _HAS_SCIPY = False

def target_likelihood_categorical(X, y, cat_cols=None, positive_label='Y', min_count=20):
    y_bin = (y == positive_label).astype(int)
    if cat_cols is None:
        cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()
    baseline = y_bin.mean()
    n_rows = len(X)

    feature_rows = []
    level_frames = []

    for col in cat_cols:
        tmp = pd.DataFrame({col: X[col], '_y': y_bin})
        grp = (
            tmp.groupby(col, dropna=False)['_y']
               .agg(n='count', events='sum')
               .reset_index()
        )
        grp['feature'] = col
        grp['rate'] = grp['events'] / grp['n']
        grp['lift'] = grp['rate'] / baseline if baseline > 0 else np.nan
        grp['share'] = grp['n'] / n_rows

        # feature-level metrics
        wstd = np.sqrt(np.average((grp['rate'] - baseline) ** 2, weights=grp['n']))
        rate_range = grp['rate'].max() - grp['rate'].min()
        low_support_frac = (grp['n'] < min_count).mean()

        if _HAS_SCIPY:
            try:
                ct = pd.crosstab(X[col], y_bin)
                _, pval, _, _ = chi2_contingency(ct)
            except Exception:
                pval = np.nan
        else:
            pval = np.nan

        feature_rows.append({
            'feature': col,
            'levels': int(grp.shape[0]),
            'baseline_rate': baseline,
            'weighted_rate_std': wstd,
            'rate_range': rate_range,
            'low_support_frac': low_support_frac,
            'chi2_p': pval
        })

        level_frames.append(
            grp[['feature', col, 'n', 'events', 'rate', 'lift', 'share']]
              .sort_values('lift', ascending=False)
        )

    feature_summary = (
        pd.DataFrame(feature_rows)
          .sort_values(['weighted_rate_std', 'rate_range'], ascending=[False, False])
          .reset_index(drop=True)
    )
    per_level = pd.concat(level_frames, ignore_index=True)

    return feature_summary, per_level

# Example usage on training data
cat_summary, cat_levels = target_likelihood_categorical(X_train, y_train, positive_label='Y', min_count=20)
display(cat_summary)
display(cat_levels.head(50))
```

#### **4.4.2** Explore the relationships between numerical features and the target variable to understand their impact on the target outcome using appropriate visualisation techniques to identify trends and potential interactions. <font color="red">[5 Marks]</font>

```python
# Visualise the relationship between numerical features and the target variable to understand their impact on the target outcome

# Numeric columns and binary target
num_cols = X_train.select_dtypes(include=[np.number]).columns.tolist()
y_bin = (y_train == 'Y').astype(int)
baseline = y_bin.mean()

# 1) Point-biserial correlation (numeric vs binary target)
corr = X_train[num_cols].corrwith(y_bin)
order = corr.abs().sort_values(ascending=False).index

plt.figure(figsize=(10, max(4, 0.35*len(num_cols))))
sns.barplot(x=corr[order], y=order, orient='h', palette='vlag')
plt.axvline(0, color='k', lw=1)
plt.title('Correlation of numerical features with target (Y=1)')
plt.xlabel('Correlation')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

# 2) Target rate vs quantile bins for top K numerical features
top_k = min(9, len(num_cols))
top_cols = order[:top_k]

ncols = 3
nrows = int(np.ceil(top_k / ncols))
fig, axes = plt.subplots(nrows, ncols, figsize=(4.8*ncols, 3.6*nrows), squeeze=False)

for i, col in enumerate(top_cols):
    ax = axes[i // ncols, i % ncols]
    bins = pd.qcut(X_train[col], q=10, duplicates='drop')
    tmp = pd.DataFrame({'bin': bins, 'y': y_bin})
    grp = tmp.groupby('bin', dropna=False)['y'].agg(rate='mean', n='count').reset_index()

    sns.barplot(data=grp, x='bin', y='rate', ax=ax, color='#4C72B0')
    ax.axhline(baseline, color='red', linestyle='--', linewidth=1)
    ax.set_title(f'{col}  (n bins={len(grp)})')
    ax.set_xlabel('')
    ax.set_ylabel('Target rate')
    for tick in ax.get_xticklabels():
        tick.set_rotation(60)

# Remove empty subplots
for j in range(i + 1, nrows * ncols):
    fig.delaxes(axes[j // ncols, j % ncols])

plt.tight_layout()
plt.show()
```

```python
# Create new features based on your understanding for both training and validation data

def add_features(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()

    # Date/time features
    if {'policy_bind_date', 'incident_date'}.issubset(out.columns):
        out['policy_age_days'] = (out['incident_date'] - out['policy_bind_date']).dt.days.clip(lower=0)
        out['policy_bind_year'] = out['policy_bind_date'].dt.year
        out['policy_bind_month'] = out['policy_bind_date'].dt.month
        out['incident_month'] = out['incident_date'].dt.month
        out['incident_dow'] = out['incident_date'].dt.dayofweek
        out['incident_is_weekend'] = out['incident_date'].dt.dayofweek.isin([5, 6]).astype(int)

    # Cyclical transform of incident hour
    if 'incident_hour_of_the_day' in out.columns:
        h = out['incident_hour_of_the_day'] % 24
        out['hour_sin'] = np.sin(2 * np.pi * h / 24)
        out['hour_cos'] = np.cos(2 * np.pi * h / 24)

    # Claim composition ratios
    if 'total_claim_amount' in out.columns:
        for comp in ['injury_claim', 'property_claim', 'vehicle_claim']:
            if comp in out.columns:
                r = out[comp] / out['total_claim_amount'].replace(0, np.nan)
                out[f'{comp}_ratio'] = r.fillna(0)

    # Per-vehicle and premium ratios
    if {'total_claim_amount', 'number_of_vehicles_involved'}.issubset(out.columns):
        denom = out['number_of_vehicles_involved'].replace(0, np.nan)
        out['claim_per_vehicle'] = (out['total_claim_amount'] / denom).fillna(out['total_claim_amount'])
    if {'total_claim_amount', 'policy_annual_premium'}.issubset(out.columns):
        out['claim_to_premium'] = (out['total_claim_amount'] / out['policy_annual_premium'].replace(0, np.nan)).fillna(0)
        out['premium_to_claim'] = (out['policy_annual_premium'] / out['total_claim_amount'].replace(0, np.nan)).fillna(0)

    # Indicators and simple interactions
    if 'umbrella_limit' in out.columns:
        out['has_umbrella'] = (out['umbrella_limit'] > 0).astype(int)
    if 'age' in out.columns:
        out['age_decade'] = (out['age'] // 10) * 10
    if {'injury_claim', 'bodily_injuries'}.issubset(out.columns):
        out['injury_per_injury'] = out['injury_claim'] / (out['bodily_injuries'] + 1)
    if 'witnesses' in out.columns:
        out['has_witness'] = (out['witnesses'] > 0).astype(int)
    if {'capital-gains', 'capital-loss'}.issubset(out.columns):
        out['net_capital'] = out['capital-gains'] + out['capital-loss']

    # Unknown-category flags (if present)
    for col in ['property_damage', 'police_report_available', 'collision_type', 'authorities_contacted']:
        if col in out.columns:
            out[f'{col}_unknown'] = (out[col].astype(str).str.lower() == 'unknown').astype(int)

    return out

# Apply to train and validation
X_train_fe = add_features(X_train)
X_val_fe = add_features(X_val)
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information

def combine_rare_categories(X_train, X_val, cols=None, min_count=25, other_label='Other'):
    Xtr, Xva = X_train.copy(), X_val.copy()
    if cols is None:
        cols = Xtr.select_dtypes(include=['object', 'category']).columns.tolist()

    summary = {}
    n = len(Xtr)

    for c in cols:
        vc = Xtr[c].value_counts(dropna=False)
        keep = set(vc[vc >= min_count].index)
        # always keep explicit Unknown bucket if present
        if 'Unknown' in vc.index:
            keep.add('Unknown')

        Xtr[c] = Xtr[c].where(Xtr[c].isin(keep), other_label)
        Xva[c] = Xva[c].where(Xva[c].isin(keep), other_label)

        summary[c] = {
            'kept_levels': sorted(map(str, keep))[:10],
            'n_levels_before': int(vc.size),
            'n_levels_after': int(Xtr[c].nunique(dropna=False))
        }

    print("Combined rare categories to", repr(other_label), "for columns:", list(summary.keys()))
    return Xtr, Xva, summary

cat_cols = X_train_sel.select_dtypes(include=['object', 'category']).columns.tolist()
X_train_grp, X_val_grp, rare_summary = combine_rare_categories(
    X_train_sel, X_val_sel, cols=cat_cols, min_count=25, other_label='Other'
)

# Quick check 
for c in cat_cols[:8]:
    print(f"{c}: {X_train_sel[c].nunique()} -> {X_train_grp[c].nunique()}")
```

#### **7.3.1** Plot ROC Curve  to visualise the trade-off between true positive rate and false positive rate across different classification thresholds <font color="red">[2 Marks]</font>

```python
# Import libraries or function to plot the ROC curve

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
# Define ROC function

def plot_roc(y_true, y_score, title='ROC Curve'):
    fpr, tpr, thr = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)
    youden_j = tpr - fpr
    opt_idx = np.argmax(youden_j)
    opt_cutoff = thr[opt_idx]

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.3f}')
    plt.plot([0, 1], [0, 1], 'k--', lw=1)
    plt.scatter(fpr[opt_idx], tpr[opt_idx], c='red', label=f'Best cutoff={opt_cutoff:.3f}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title(title)
    plt.legend(loc='lower right')
    plt.tight_layout()
    plt.show()

    return {'fpr': fpr, 'tpr': tpr, 'thr': thr, 'auc': roc_auc, 'opt_cutoff': float(opt_cutoff)}
```
